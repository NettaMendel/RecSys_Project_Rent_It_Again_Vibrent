```python
# Calculate the rental period duration in days
orders['rentalPeriod'] = (orders['rentalPeriod.end'] - orders['rentalPeriod.start']).dt.days
```

```python
# Plot the distribution of the rental period
plt.figure(figsize=(10, 5))
sns.histplot(orders['rentalPeriod'], bins=range(0, orders['rentalPeriod'].max() + 1, 1), kde=True)
plt.title('Distribution of Rental Period')
plt.xlabel('Rental Period (days)')
plt.ylabel('Frequency')
plt.xticks(range(0, orders['rentalPeriod'].max() + 1, 10),rotation=90)
plt.xlim(left=0)
plt.show()
```

```python
# Plot the distribution of the start date of the rental period
plt.figure(figsize=(10, 5))
plt.hist(orders['rentalPeriod.start'], bins=10, edgecolor='black')
plt.title('Distribution of Rental Period Start Date')
plt.xlabel('Start Date')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
outfits=pd.read_parquet('../archive/data/outfits.parquet',engine='pyarrow')
outfits.head()
```

```python
# Define the rental period categories
def categorize_duration(days):
    if days <= 7:
        return 'Week'
    elif days <= 30:
        return 'Month'
    elif days <= 60:
        return '2 Months'
    else:
        return 'Other'

# Apply the categorization
orders['rentalPeriod.category'] = orders['rentalPeriod'].apply(categorize_duration)

# Plot the correlation between the rental period category and the start date
plt.figure(figsize=(10, 6))
for category in orders['rentalPeriod.category'].unique():
    subset = orders[orders['rentalPeriod.category'] == category]
    plt.scatter(subset['rentalPeriod.start'], subset['rentalPeriod.duration'], label=category)

plt.title('Correlation between Rental Period Duration and Start Date')
plt.xlabel('Rental Period Start Date')
plt.ylabel('Rental Period Duration (days)')
plt.legend(title='Rental Period Category')
plt.grid(True)
plt.show()
```

```python
# Merge with outfits on the id column
merged_df = pd.merge(orders, outfits, left_on='outfit.id',right_on='id')
merged_df.columns
```

```python
merged_df.drop(columns=['id'],inplace=True)
```

```python
# Calculate the total amount paid according to pricePerWeek and pricePerMonth
merged_df['totalPaidPerWeek'] = (merged_df['rentalPeriod'] / 7) * merged_df['pricePerWeek']
merged_df['totalPaidPerMonth'] = (merged_df['rentalPeriod'] / 30) * merged_df['pricePerMonth']
```

```python
# Create histogram for totalPaidPerWeek and totalPaidPerMonth in the same plot
plt.figure(figsize=(10, 5))
plt.hist(merged_df['totalPaidPerWeek'], bins=30, alpha=0.7, color='blue', label='Total Paid Per Week')
plt.hist(merged_df['totalPaidPerMonth'], bins=30, alpha=0.7, color='green', label='Total Paid Per Month')
plt.title('Distribution of Total Amount Paid According to Price Per Week and Price Per Month')
plt.xlabel('Total Amount Paid (NOK)')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

It is look like it is better to rent in the monthly rate

```python
# Sort the dataframe by 'rentalPeriod.start'
merged_df = merged_df.sort_values(by='rentalPeriod.start')

# Create a plot to show the trend of totalPaidPerWeek and totalPaidPerMonth over time
plt.figure(figsize=(12, 6))
plt.plot(merged_df['rentalPeriod.start'], merged_df['totalPaidPerWeek'], label='Total Paid Per Week', color='blue')
plt.plot(merged_df['rentalPeriod.start'], merged_df['totalPaidPerMonth'], label='Total Paid Per Month', color='green')
plt.title('Trend of Total Amount Paid Over Time')
plt.xlabel('Rental Period Start Date')
plt.ylabel('Total Amount Paid')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Aggregate the data by month to smooth the plot
df=merged_df.set_index('rentalPeriod.start')
monthly_data = df.resample('M').sum()

# Create a plot to show the trend of totalPaidPerWeek and totalPaidPerMonth over time
plt.figure(figsize=(12, 6))
plt.plot(monthly_data.index, monthly_data['totalPaidPerWeek'], label='Total Paid Per Week', color='blue')
plt.plot(monthly_data.index, monthly_data['totalPaidPerMonth'], label='Total Paid Per Month', color='green')

# Add trend lines
z_week = np.polyfit(monthly_data.index.astype(int), monthly_data['totalPaidPerWeek'], 1)
p_week = np.poly1d(z_week)
plt.plot(monthly_data.index, p_week(monthly_data.index.astype(int)), linestyle='--', color='blue', label='Trend Line Per Week')

z_month = np.polyfit(monthly_data.index.astype(int), monthly_data['totalPaidPerMonth'], 1)
p_month = np.poly1d(z_month)
plt.plot(monthly_data.index, p_month(monthly_data.index.astype(int)), linestyle='--', color='green', label='Trend Line Per Month')

plt.title('Trend of Total Amount Paid Over Time (Monthly Aggregation)')
plt.xlabel('Rental Period Start Date')
plt.ylabel('Total Amount Paid (NOK per Month)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
